# Plots Different Layers

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# File paths for the different CSVs
file_paths = [
    '/home/ale/TT_RLProject_2024/Task_2/REINFORCE_70_Logs.csv',
    '/home/ale/TT_RLProject_2024/Task_2/REINFORCE_70_64x2_Logs.csv',
    #'/home/ale/TT_RLProject_2024/Task_2/REINFORCE_70_128x2_Logs.csv',
]

# Create a figure and axis for the plot
plt.figure(figsize=(18, 6))

# Function to get the max reward in every n rows
def max_in_intervals(data, interval=10):
    return data.groupby(data.index // interval).max()

# Plot each algorithm's data with max reward every 10 rows
for i, file_path in enumerate(file_paths):
    data = pd.read_csv(file_path)
    max_rewards = max_in_intervals(data[['Episode', 'Train Reward']], interval=10)
    plt.plot(max_rewards['Episode'], max_rewards['Train Reward'], label=f'Algorithm {i+1}', markevery=1)

# Add labels and title
plt.xlabel('Episode')
plt.ylabel('Train Reward')
plt.title('Max Train Reward per 10 Episodes for Different Algorithms')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
